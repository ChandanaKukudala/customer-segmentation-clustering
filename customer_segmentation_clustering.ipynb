## K-Means Clustering

K-Means is an unsupervised machine learning algorithm used to group similar data points into clusters based on distance.

In this analysis, we use K-Means to segment individuals based on their income and spending behavior.

### Feature Selection and Scaling

We select two key features:
- Monthly Income
- Monthly Expenses

Since K-Means is distance-based, scaling is necessary to ensure both features contribute equally,  and unscaled features can bias the results.

StandardScaler is used to normalize the features so that they have a mean of 0 and a standard deviation of 1.

```python
from sklearn.preprocessing import StandardScaler

X = df[['monthly_income_usd', 'monthly_expenses_usd']]

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```


### Elbow Method

The Elbow Method is used to determine the optimal number of clusters (K) for K-Means.

It works by calculating the Within-Cluster Sum of Squares (WCSS) for different values of K.

WCSS measures how tightly grouped the data points are within each cluster.

We choose the value of K where adding more clusters does not significantly improve the model and  the decrease in WCSS starts to slow down.
This point is called the "elbow".

```python
from sklearn.cluster import KMeans

wcss = []

for i in range(1,11):
  kmeans = KMeans(n_clusters=i, random_state=42)
  kmeans.fit_transform(X_scaled)
  wcss.append(kmeans.inertia_)

plt.plot(range(1,11), wcss, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.title('Elbow Method')
plt.show()
```

### Insights from Elbow Method

From the Elbow Method graph, the optimal number of clusters appears to be 4,
as the reduction in WCSS slows down after this point.

### Applying K-Means

The K-Means model is initialized with 4 clusters based on the Elbow Method.

The `fit_predict()` function is used to:
- Fit the model to the data
- Assign each data point to a cluster

A new column `cluster` is created to store the cluster labels for each individual.

```python
kmeans = KMeans(n_clusters=4, random_state=42)
df['cluster'] = kmeans.fit_predict(X_scaled)

```

### Cluster Visualization

A scatter plot is used to visualize the clusters.

- Each point represents an individual
- X-axis shows monthly income
- Y-axis shows monthly expenses
- Colors represent different clusters assigned by K-Means

This helps in understanding how individuals are grouped based on financial behavior.

```python
plt.scatter(df['monthly_income_usd'], df['monthly_expenses_usd'], c=df['cluster'])
plt.xlabel('Monthly Income')
plt.ylabel('Monthly Expenses')
plt.title('Customer Segmentation using K.Means')
```

### Cluster Insights

The clusters show a clear relationship between income and expenses, where higher income is associated with higher spending.

K-Means has divided the data into segments along this trend, resulting in wedge-shaped clusters rather than clearly separated groups.

This indicates that the data follows a linear pattern, and clustering is segmenting individuals based on spending levels relative to income.

Cluster interpretation:
- High income, high expenses → High spenders
- Moderate income and expenses → Average consumers
- Low income, moderate expenses → Budget-conscious group
- Low income, low expenses → Low-income individualsc
